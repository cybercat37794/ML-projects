```python
# importing library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cybercat37794/ML-projects/main/Car%20Driving%20Risk%20Prediction/car%20driving%20risk%20analysis.csv')
```

```python
df
```

```python
# filling the null value of experience with mean of the experience column values
df.experience = df.experience.fillna(df.experience.mean())
```

```python
df.experience
```

```python
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['speed', 'car_age', 'experience']], df.risk)
```

```python
reg.predict([[160, 10, 7]])
```
